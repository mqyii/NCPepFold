```python
# 删除yaml文件中小分子的绝对路径
import yaml
from pathlib import Path

def update_input_paths(data, prefix_path):
    """递归地更新所有input路径."""
    if isinstance(data, dict):
        for key, value in data.items():
            if key == 'input' and isinstance(value, str) and value.startswith(prefix_path):
                data[key] = value[len(prefix_path):]
            else:
                update_input_paths(value, prefix_path)
    elif isinstance(data, list):
        for item in data:
            update_input_paths(item, prefix_path)

def update_yaml_file(file_path, prefix_path):
    """更新YAML文件中的input路径."""
    with open(file_path, 'r') as file:
        data = yaml.safe_load(file)

    update_input_paths(data, prefix_path)  # 递归更新路径

    with open(file_path, 'w') as file:
        yaml.safe_dump(data, file, sort_keys=False)

# 主程序入口
def main():
    parent_folder = Path("/home/light/mqy/ncaa/data/这次不用的/ncaa_linear_final_analysis_new/no_alignment")
    prefix_path = "/home/light/mqy/ncaa/RoseTTAFold-All-Atom/"

    # 遍历父文件夹中的所有子文件夹
    for subdir in parent_folder.iterdir():
        if subdir.is_dir():  # 确保是一个目录
            config_file = subdir / "config_aa.yaml"
            if config_file.exists():  # 检查 config 文件是否存在
                update_yaml_file(config_file, prefix_path)
                print(f"Updated inputs in {config_file}")

main()


```

```python
# sm_inputs 和 protein_inputs 是否有重复链

def extract_chain_names_from_inputs(inputs):
    chains = []
    for key, value in inputs.items():
            chains.append(key)
    return chains

def find_folders_with_duplicate_chain_names(parent_folder):
    parent_path = Path(parent_folder)
    duplicate_folders = []

    for subdir in parent_path.iterdir():
        if subdir.is_dir():
            config_file = subdir / "config_aa.yaml"
            if config_file.is_file():
                with open(config_file, 'r') as f:
                    config = yaml.safe_load(f)

                if 'protein_inputs' in config and 'sm_inputs' in config:
                    protein_chains = extract_chain_names_from_inputs(config['protein_inputs'])
                    sm_chains = extract_chain_names_from_inputs(config['sm_inputs'])

                    if set(protein_chains) & set(sm_chains):  # Check if there's any overlap
                        duplicate_folders.append(subdir.name)
    
    return duplicate_folders

parent_folder = "/home/light/mqy/ncaa/data/ncaa_linear_error/2-CH3/DA2"
duplicate_folders = find_folders_with_duplicate_chain_names(parent_folder)
for folder in duplicate_folders:
    print(f"Folder with duplicate chain names: {folder}")
```

```python
# ncaa_linear 的yaml 加中括号

def update_covale_inputs(yaml_file):
    with open(yaml_file, 'r') as file:
        config = yaml.safe_load(file)

    if 'covale_inputs' in config:
        covale_inputs = config['covale_inputs']
        # Convert string representation to list if it's a string
        if isinstance(covale_inputs, str):
            covale_inputs = eval(covale_inputs)

        # Wrap each tuple with additional brackets
        if isinstance(covale_inputs, list):
            wrapped_inputs = [[ci] for ci in covale_inputs]
            wrapped_inputs = [[wrapped_input] for wrapped_input in wrapped_inputs]

            # Convert back to string with correct quote style
            quote_style = '"' if '"' in str(config['covale_inputs']) else "'"
            covale_inputs_str = repr(wrapped_inputs)
            if quote_style == '"':
                covale_inputs_str = covale_inputs_str.replace("'", '"')
            else:
                covale_inputs_str = covale_inputs_str.replace('"', "'")

            # Update the config with the correctly quoted string
            config['covale_inputs'] = covale_inputs_str
    else:
        raise ValueError("No 'covale_inputs' found in the provided yaml file")

    with open(yaml_file, 'w') as file:
        yaml.safe_dump(config, file, default_flow_style=False)

def process_yaml_files(parent_folder):
    parent_path = Path(parent_folder)

    for subdir in parent_path.iterdir():
        if subdir.is_dir():
            yaml_file = subdir / "config_aa.yaml"
            if yaml_file.exists():
                with open(yaml_file, 'r') as file:
                    config = yaml.safe_load(file)
                if 'covale_inputs' in config:
                    covale_inputs = str(config['covale_inputs'])
                    if "[[[" in covale_inputs:
                        continue
                    else:
                        print(subdir.name)
                        update_covale_inputs(yaml_file)
                        print(f"Updated {yaml_file}")
            else:
                print(f"{subdir.name}的yaml不存在")

# 使用示例
parent_folder = "/home/light/mqy/ncaa/data/test_set/ncaa_cyclic/monomer/ss/other"
process_yaml_files(parent_folder)
```

```python
from pathlib import Path
from Bio.PDB.PDBParser import PDBParser
from Bio.PDB.Polypeptide import PPBuilder
from Bio.PDB.Residue import Residue
from Bio.PDB.Atom import Atom
from Bio.PDB.Structure import Structure
from Bio.PDB.Chain import Chain
from Bio.PDB.Model import Model
from Bio.PDB.PDBIO import PDBIO
from Bio.PDB.Entity import Entity
from Bio.PDB import PDBParser, Select, PDBIO
import re
import warnings
from loguru import logger

warnings.filterwarnings("ignore")

standard_amino_acids = [
    "ALA",
    "CYS",
    "ASP",
    "GLU",
    "PHE",
    "GLY",
    "HIS",
    "ILE",
    "LYS",
    "LEU",
    "MET",
    "ASN",
    "PRO",
    "GLN",
    "ARG",
    "SER",
    "THR",
    "VAL",
    "TRP",
    "TYR",
]


def is_aa(aa_str):
    if aa_str in standard_amino_acids:
        return True
    else:
        return False


class NonHydrogenSelect(Select):
    """只选择非氢原子的类，使用正则表达式匹配氢原子"""

    def accept_atom(self, atom):
        # 匹配 'H' 或以数字开头后跟 'H' 的原子名称
        flag = True
        if re.match(r"^\d*H", atom.get_id()):
            flag = False
        if re.match("OXT", atom.get_id()):
            flag = False
        return flag


class NoWatersSelect(Select):
    """选择器类，用于排除水分子。"""

    def accept_residue(self, residue):
        # 只接受非水分子的残基
        return residue.get_resname() not in ["HOH", "WAT"]


def remove_water(structure):
    # 遍历每个模型、每个链、每个残基，收集水分子的位置
    water_residues = []
    for model in structure:
        for chain in model:
            for residue in chain:
                if residue.get_resname() in ["HOH", "WAT"]:
                    water_residues.append((model.id, chain.id, residue.id))
    # 在结构中移除这些水分子
    for model_id, chain_id, res_id in water_residues:
        model = structure[model_id]
        chain = model[chain_id]
        del chain[res_id]
    logger.debug(f"Water molecules have been removed from the {str(structure)} in memory.")

def remove_ion(structure):
    # 遍历每个模型、每个链、每个残基，收集水分子的位置
    water_residues = []
    for model in structure:
        for chain in model:
            for residue in chain:
                if residue.get_resname() in ["ZN", "CD", "MG", "SO4", "NI", "CL", "CA", "PO4", "NA", "UNX","FAR", "ACT", "ANS", "EDO", "GOL", "K", "TBF", "NME", "BFC", "SIN", "ACT", "K", "NAG"]:
                    water_residues.append((model.id, chain.id, residue.id))
    # 在结构中移除这些水分子
    for model_id, chain_id, res_id in water_residues:
        model = structure[model_id]
        chain = model[chain_id]
        del chain[res_id]
    logger.debug(f"Zn molecules have been removed from the {str(structure)} in memory.")

def remove_dna(structure):
    # 遍历每个模型、每个链、每个残基，收集水分子的位置
    water_residues = []
    for model in structure:
        for chain in model:
            for residue in chain:
                if residue.get_resname() in ["A", "U", "C", "G", "T"]:
                    water_residues.append((model.id, chain.id, residue.id))
    # 在结构中移除这些水分子
    for model_id, chain_id, res_id in water_residues:
        model = structure[model_id]
        chain = model[chain_id]
        del chain[res_id]
    logger.debug(f"Zn molecules have been removed from the {str(structure)} in memory.")

def get_pos(chain, ncaa):
    print(chain)

```

```python
# 查看 pep.fasta 和 protein-x.fasta 是否只有一条链

def check_fasta_chains(fasta_file):
    with open(fasta_file, 'r') as f:
        records = list(SeqIO.parse(f, "fasta"))
        return len(records) == 1

def check_subfolders(base_dir):
    for subdir in base_dir.iterdir():
        if subdir.is_dir():
            pep_fasta = subdir / f"{subdir.name[:-1]}-pep-aa.fasta"
            protein_fastas = list(subdir.glob(f"{subdir.name[:-1]}-protein-*.fasta"))
            all_single_chain = True
            
            if pep_fasta.exists() and not check_fasta_chains(pep_fasta):
                print(f"{pep_fasta} contains more than one chain.")
                all_single_chain = False
            
            for protein_fasta in protein_fastas:
                if not check_fasta_chains(protein_fasta):
                    print(f"{protein_fasta} contains more than one chain.")
                    all_single_chain = False

base_dir = Path("/home/light/mqy/ncaa/data/ncaa_linear_error/2-CH3/M3L")
check_subfolders(base_dir)
```

```python


def get_shortest_chain(structure):
    min_length = float('inf')
    shortest_chain = None
    for chain in structure.get_chains():
        chain_length = len(list(chain.get_residues()))
        if chain_length < min_length:
            min_length = chain_length
            shortest_chain = chain
    return shortest_chain

def compare_atoms(residue, standard_atoms):
    residue_atoms = [atom.get_name() for atom in residue if atom.element != 'H']
    return set(residue_atoms) == set(standard_atoms)

def process_pdb_files(base_dir, amino_acid_atoms):
    for subdir in base_dir.iterdir():
        if subdir.is_dir():
            pdb_file = subdir / f"{subdir.name}_clean.pdb"
            if pdb_file.exists():
                parser = PDBParser(QUIET=True)
                structure = parser.get_structure(subdir.name, pdb_file)
                shortest_chain = get_shortest_chain(structure)
                if shortest_chain:
                    for i, residue in enumerate(shortest_chain.get_residues(), start=1):
                        residue_name = residue.get_resname()
                        if residue_name in amino_acid_atoms:
                            if not compare_atoms(residue, amino_acid_atoms[residue_name]):
                                print(f"{subdir.name}: Residue {i} {residue_name} in PDB has different atoms.")
                                print("dict:", amino_acid_atoms[residue_name])
                                a = []
                                for atom in residue.get_atoms():
                                    a.append(atom.get_name())
                                print("pdb: ", a)
                                print()
                        else:
                            print(f"{subdir.name}: Residue {i} {residue_name} in PDB is not in the amino acid dictionary.")

# 设置父文件夹路径
base_dir = Path("/home/light/mqy/ncaa/data/ncaa_linear_error/2-CH3/MLY")

# 执行比较操作
process_pdb_files(base_dir, amino_acid_atoms)
```
